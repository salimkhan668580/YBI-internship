**Source**

This dataset was taken from the StatLib library which is maintained at Carnegie Mellon University. THe datadet was used in the 1983 American Statistical Association Exposition

**Data Set Information**

This dataset is a slightly modified version of the dataset provided in the StatLib library .In line with the use by Ross Quinlan (1993) in predicting the attirbute "mpg",8 of the original instances were removed beacuse they had unknown for the "mpg" attribute the orignal dataset is avilable in the "auot-mpg.data-orignal".

"The data concerns city-cycle fuel consumption in miles per gallon, to be predicted in term ogf 3 multivalued dixvrete and 5 continious attribute ".(Quinian,1993)

**Attribute information:**
1. mpg: continuous
2. cylinders: multi-valued discrete
3. displacement: continuous
4. horsepower: continuous
5. weigth:continuous
6. acceleration:continuous
7. model year:multi-valued discrete
8. origin:multi-valued discrete
9. car name: string(unique for each instance)

**Define Target Variable (y) and Feature Variables (X)**

```python
y=df['mpg']
```

```python
y.shape
```

```python
x=df[[ 'displacement', 'horsepower', 'weight',
       'acceleration']]
```

```python
x.shape
```

```python
x
```

```python
from sklearn.preprocessing import StandardScaler
ss= StandardScaler()
```

```python
x=ss.fit_transform(x)
```

```python
x
```

```python
pd.DataFrame(x).describe()
```

```python
pd.DataFrame(x).describe()
```

**Train Test Split**

```python
from sklearn.model_selection import train_test_split
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(x,y, train_size=0.7, random_state=2529)
```

```python

X_train.shape, X_test.shape, y_train.shape, y_test.shape
```

**Modeling**

```python
from sklearn.linear_model import LinearRegression
```

```python
lr=LinearRegression()
```

**Model Evaluation**

```python
# Step 6 : train or fit model
lr.fit(X_train,y_train)
```

**Model Evaluation**

```python
lr.intercept_
```

```python
lr.coef_
```

**Prediction**

```python

y_pred = lr.predict(X_test)
```

```python
y_pred
```

```python
from sklearn.metrics import mean_absolute_error,mean_absolute_percentage_error,r2_score
```

```python
mean_absolute_error(y_test,y_pred)
```

```python
mean_absolute_percentage_error(y_test,y_pred)
```

```python
r2_score(y_test,y_pred)
```

```python
from sklearn.preprocessing import PolynomialFeatures
```

```python
poly=PolynomialFeatures(degree=2,interaction_only=True,include_bias=False)
```

```python
X_train2=poly.fit_transform(X_train)
```

```python
X_test2=poly.fit_transform(X_test)
```

```python
lr.fit(X_train2,y_train)
```

```python
lr.intercept_
```

```python
lr.coef_
```

```python
y_pred_poly=lr.predict(X_test2)
```

```python
from sklearn.metrics import mean_absolute_error,mean_absolute_percentage_error,r2_score
```

```python
mean_absolute_error(y_test,y_pred_poly)
```

```python
mean_absolute_percentage_error(y_test,y_pred_poly)
```

```python
r2_score(y_test,y_pred_poly)
```

**Explaination**

This model have mean_absolute_error 3.3286968643244106 using LinearRegression and  using PolynomialFeatures the mean_absolute_error is 2.7887147720295977 so the polynomialFeatures is more efficinet
